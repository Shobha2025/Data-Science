```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customers_data.csv")
df.head()
```

# 1. Bar Plot: Top 5 Countries by Customer Count

```python
top_countries = df['Country'].value_counts().nlargest(5).reset_index()
top_countries.columns = ['Country', 'Count']
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=top_countries, x='Country', y='Count', palette='pastel')
plt.title('Top 5 Countries by Customer Count')
plt.ylabel('Customers')
plt.show()
```

# Count Plot: Number of Customers per City (Top 10)

```python
top_cities = df['City'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=df[df['City'].isin(top_cities)], x='City', order=top_cities, palette='muted')
plt.title('Customer Count by Top 10 Cities')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Line Plot: Monthly Subscriptions Over Time

```python
df['Subscription Date'] = pd.to_datetime(df['Subscription Date'], errors='coerce')
df = df.dropna(subset=['Subscription Date'])
df['Month'] = df['Subscription Date'].dt.to_period('M').astype(str)
monthly = df.groupby('Month').size().reset_index(name='Subscriptions')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly, x='Month', y='Subscriptions', marker='o', color='teal')
plt.xticks(rotation=45)
plt.title('Monthly Customer Subscriptions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Box Plot: Subscription Date Distribution by Country

```python
plt.figure(figsize=(12, 6))
df['Year'] = df['Subscription Date'].dt.year
sns.boxplot(data=df, x='Country', y='Year', palette='Set2')
plt.title('Subscription Year Distribution by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Heatmap: Missing Data Overview

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Data Heatmap')
plt.show()
```
